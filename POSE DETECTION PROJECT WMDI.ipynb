```python
#Step 1: import all necessary libraries
import cv2
import mediapipe as mp
```

```python
#Step 2: identify webcam
cap = cv2.VideoCapture(0) # Local Webcam (0), External Webcam(1)
```

```python
#Leveraging the mediapipe library used for pose detection
mpPose = mp.solutions.pose
pose= mpPose.Pose()
#Pose=mpPose(static_image_mode=False,upper_body_only=False, smooth_Landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)

#To Draw and connect the landmarks
mpDraw = mp.solutions.drawing_utils
```

```python
#Switch on your webcam

while True:
    _, img = cap.read()
    
    #Convert video/image from BGR TO RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #Apply mediapipe pose detection module for detection
    results = pose.process(imgRGB)
    print(results.pose_landmarks)
    
    #Draw Landmarks
    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
        
    cv2.putText(img, "Macdonald's Pose Detection @10Alytics",(10,50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,255), 2)
    
    
    cv2.imshow("Macdonald's Pose Detection Program", img)
    if cv2.waitKey(0) & 0xFF == ord('q'):
        break
#Release the capture once all the processing is done.
cao.release()
cv2.destroyAllWindows()
```
